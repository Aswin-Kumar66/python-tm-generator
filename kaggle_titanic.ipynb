```python
import pandas as pd
from google.colab import files
import io

uploaded = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded['train.csv']))
df.head()
```

```python
df.isna().sum()
```

```python
df.shape
```

```python

df = df.replace('female' , 0 )
df = df.replace('male' , 1 )
```

```python
df.head()
```

```python
df = df.drop('Name' , axis = 1 )
df.head()
```

```python
df = df.drop('Ticket' ,  axis = 1)
df.head()
```

```python
df['Age'].describe()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())

df.head()
```

```python
df.isna().sum()
```

```python
df = df.drop("Cabin" , axis = 1)
```

```python
df.head()
```

```python
df['Embarked'].isna().sum()
```

```python
df = df.dropna(subset = "Embarked")

df['Embarked'].unique()
```

```python
df = df.replace('S', 0)
df = df.replace('C', 0.5)
df = df.replace('Q', 1)
df.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['Age'] = scaler.fit_transform(df[['Age']])
df['Fare'] = scaler.fit_transform(df[['Fare']])
df['Parch'] = scaler.fit_transform(df[['Parch']])
df['SibSp'] = scaler.fit_transform(df[['SibSp']])
df['Pclass'] = scaler.fit_transform(df[['Pclass']])
```

```python
df.head()
df.describe()
df = df.drop('PassengerId' , axis = 1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.drop('Survived' , axis = 1), df.Survived, test_size=0.25, random_state=0)
```

```python
logisticRegr = LogisticRegression()
logisticRegr.fit(x_train, y_train)

score = logisticRegr.score(x_test, y_test)
print(score)
```

```python
from sklearn.tree import DecisionTreeClassifier
decisiontree = DecisionTreeClassifier()
decisiontree.fit(x_train, y_train)

score = decisiontree.score(x_test, y_test)
print(score)
```

```python
from sklearn.ensemble import RandomForestClassifier
randomforest = RandomForestClassifier(n_estimators = 600)
randomforest.fit(x_train, y_train)

score = randomforest.score(x_test, y_test)
print(score)
```

```python
from sklearn.svm import SVC
svm = SVC(kernel="rbf", gamma=0.5, C=1.0)
svm.fit(x_train, y_train)

score = svm.score(x_test, y_test)
print(score)
```
